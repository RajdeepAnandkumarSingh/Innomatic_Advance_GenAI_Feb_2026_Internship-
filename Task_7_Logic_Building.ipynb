```python
#1
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.is_on = False  # Initial state is OFF

    def process_action(self, action):
        # Convert to uppercase to handle 'on', 'On', or 'ON'
        if action.upper() == "ON":
            self.is_on = True
        elif action.upper() == "OFF":
            self.is_on = False

    def display_status(self):
        status = "ON" if self.is_on else "OFF"
        print(f"{self.name} is {status}")

light_name = "Bedroom Light"
action_input = "ON"

my_light = SmartLight(light_name)
my_light.process_action(action_input)
my_light.display_status()
```

```python
#2
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

name_input = "Rahul"
id_input = "EMP102"
dept_input = "IT"

new_employee = Employee(name_input, id_input, dept_input)
new_employee.display_id_card()
```

```python
#3
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def save_and_display(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

name_input = "Anita"
phone_input = "9876543210"

user_contact = Contact(name_input, phone_input)
user_contact.save_and_display()
```

```python
#4
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):

        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

product_name = "Headphones"
product_price = 2499

item = Product(product_name, product_price)
item.print_price_tag()
```

```python
#5
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating}")

movie_name = "Inception"
movie_rating = 4.8

my_movie = Movie(movie_name, movie_rating)
my_movie.display_details()
```

```python
#6
class Delivery:
    def __init__(self, name, address):
        self.name = name
        self.address = address

    def print_details(self):
        print("Delivery Details")
        print(f"Customer: {self.name}")
        print(f"Address: {self.address}")

customer_name = "Suman"
delivery_address = "Hyderabad"

shipment = Delivery(customer_name, delivery_address)
shipment.print_details()
```
