```python
#2

employees = {
    "ravi":75000,
    "ankita":68000,
    "kiran":72000
}
higest_earner = max(employees,key=employees.get)
higest_salary = employees[higest_earner]
print(f"higest salary:{higest_earner}-{higest_salary}")
```

```python
#3

numbers=[45,22,89,10,66]
max_val = numbers[0]
min_val = numbers[0]

for num in numbers:
  if num > max_val:
    max_val = num
  if num < min_val:
    min_val = num

print("List:",numbers)
print("max :",max_val)
print("min :",min_val)
```

```python
#7

text ="pythonp"
char_count ={}

for char in text:
  if char in char_count:
    char_count[char] += 1
  else:
    char_count[char] =1
print(char_count)
```
