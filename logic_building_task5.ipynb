```python
#1
def analyze_engagement(likes_data):
    total_likes = 0

    for count in likes_data:
        total_likes += count

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

# Example usage
post_likes = [400, 350, 300, 100]
analyze_engagement(post_likes)
```

```python
#2
def monitor_medicine_stock(current_stock):
    if current_stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print(f"Medicine Stock: {current_stock}")
    print(f"Status: {status}")

monitor_medicine_stock(8)
```

```python
#3
def analyze_rainfall(rainfall_data, required_level):
    total_rainfall = 0
    count = 0

    for amount in rainfall_data:
        total_rainfall += amount
        count += 1

    if count == 0:
        return "No data available"

    average_rainfall = total_rainfall / count

    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print(f"Average Rainfall: {average_rainfall}")
    print(f"Rainfall Status: {status}")

# Example usage
seasonal_data = [80, 65, 90, 53]
required_threshold = 70
analyze_rainfall(seasonal_data, required_threshold)
```

```python
#4
def check_duplicate_usernames(usernames):
    unique_usernames = set()
    duplicate_found = False

    for user in usernames:
        if user in unique_usernames:
            duplicate_found = True
            break
        unique_usernames.add(user)

    if duplicate_found:
        status = "Yes"
    else:
        status = "No"

    print(f"Duplicate Accounts Found: {status}")

# Example usage
user_list = ["user123", "admin_01", "user123", "dev_spark"]
check_duplicate_usernames(user_list)
```

```python
#6
def filter_premium_crops(crop_prices):
    premium_list = []

    for price in crop_prices:
        if price > 2000:
            premium_list.append(price)

    print(f"Premium Crops: {premium_list}")

# Example usage
current_market_prices = [1500, 2500, 1800, 3200, 2000]
filter_premium_crops(current_market_prices)
```

```python
#7
def evaluate_application_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {error_count}")
    print(f"Application Health: {status}")

# Example usage
evaluate_application_health(3)
```

```python
#8
def validate_transaction(amount):
    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print(f"Transaction Amount: ₹{amount}")
    print(f"Status: {status}")

# Example usage
validate_transaction(55000)
```

```python
#9
def check_exam_eligibility(attendance_records):
    total_sessions = len(attendance_records)
    present_count = 0

    for record in attendance_records:
        if record == "Present":
            present_count += 1

    if total_sessions == 0:
        return "No records found"

    attendance_percentage = (present_count / total_sessions) * 100

    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print(f"Attendance Percentage: {attendance_percentage}%")
    print(f"Eligibility Status: {status}")

# Example usage (P = Present, A = Absent)
records = ["Present", "Present", "Present", "Present", "Present", "Absent"]
check_exam_eligibility(records)
```

```python
#10
def calculate_electricity_bill(units):
    bill_amount = 0

    if units <= 100:
        bill_amount = units * 3
    elif units <= 200:
        bill_amount = (100 * 3) + ((units - 100) * 5)
    else:
        bill_amount = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill_amount < 500:
        usage_status = "Low Usage"
    elif 500 <= bill_amount <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    return bill_amount, usage_status

# Real-Time Use Case: Smart Meter Processing
units_consumed = 250
bill, status = calculate_electricity_bill(units_consumed)

print(f"Total Bill: ₹{bill}")
print(f"Usage Level: {status}")
```
