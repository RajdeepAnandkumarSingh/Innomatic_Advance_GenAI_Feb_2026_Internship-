```python
#1
def validate_recharge(amount):
    """Checks if the amount meets the telecom provider's rules."""
    valid_plans = [199, 299, 399, 599]


    if amount < 50:
        return False, "Amount must be at least ₹50."

    if amount not in valid_plans:
        return False, f"Invalid plan. Please choose from {valid_plans}."

    return True, "Validation successful!"

def start_recharge_process():
    """Handles the user input and retry logic."""
    print("--- Welcome to Telecom Recharge ---")

    while True:
        try:
            user_input = int(input("\nEnter recharge amount (₹): "))
            is_valid, message = validate_recharge(user_input)

            if is_valid:
                print(f"Success: {message}")
                print(f"Processing payment for ₹{user_input}...")
                break  # Exit the loop on success
            else:
                print(f" Error: {message} Please try again.")

        except ValueError:
            print(" Invalid input! Please enter a numerical value.")

# Run the program
start_recharge_process()
```

```python
#2
def check_inventory(stock_data):
    """Checks each product and prints a status alert."""
    print(f"{'Product':<15} | {'Stock':<10} | {'Status'}")
    print("-" * 40)

    for product, quantity in stock_data.items():
        if quantity < 15:
            status = " Reorder Alert"
        else:
            status = " Stock OK"

        print(f"{product:<15} | {quantity:<10} | {status}")

inventory = {
    "Laptops": 25,
    "Mice": 12,
    "Keyboards": 8,
    "Monitors": 20,
    "USB Cables": 5
}

check_inventory(inventory)
```

```python
#3
def check_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

student_marks = [60, 45, 70, 55, 40]

result = check_result(student_marks)
print("Result:", result)
```

```python
#4
def calculate_fare(km, is_peak_hour):
    base_fare = 50
    rate_per_km = 12
    total_fare = base_fare + (km * rate_per_km)

    if is_peak_hour:
        total_fare *= 1.25

    return total_fare

def cab_service():
    while True:
        user_input = input("Enter distance in km (or 'exit' to quit): ").lower()
        if user_input == 'exit':
            break

        try:
            dist = float(user_input)
            peak_input = input("Peak hour? (yes/no): ").lower()
            is_peak = peak_input == 'yes'

            fare = calculate_fare(dist, is_peak)
            print(f"Total Fare: ₹{fare:.2f}")
        except ValueError:
            print("Invalid input. Please enter a number.")

cab_service()
```

```python
#5
def check_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    for status in attendance_list:
        if status == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

while True:
    user_input = input("Enter attendance (P/A separated by space) or 'exit': ").upper()
    if user_input == "EXIT":
        break

    data = user_input.split()
    if not data:
        continue

    result = check_eligibility(data)
    print(f"Status: {result}")
```

```python
#6
def check_password_strength(password):
    if len(password) < 8:
        return "Weak: Minimum 8 characters required"

    has_digit = False
    has_special = False
    special_chars = "@#$"

    for char in password:
        if char.isdigit():
            has_digit = True
        elif char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    elif not has_digit:
        return "Weak: Must contain at least one digit"
    else:
        return "Weak: Must contain at least one special character (@#$)"

while True:
    pwd = input("\nEnter password to check (or 'quit' to exit): ")
    if pwd.lower() == 'quit':
        break

    result = check_password_strength(pwd)
    print(result)
```
