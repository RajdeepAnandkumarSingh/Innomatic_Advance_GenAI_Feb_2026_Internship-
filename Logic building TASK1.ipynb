```python
#1
stored_username = "admin"
stored_password = "1234"

username_input = input("Enter username: ")
password_input = input("Enter password: ")

if username_input == stored_username and password_input == stored_password:
    print("Login successful!")

else:
    print("Invalid username or password.")
```

```python
#2
marks = [45,78,90,33,60]
pass_count = 0
fail_count = 0

PASS_MARK = 50
for score in marks:
    if score >= PASS_MARK:
        pass_count += 1
    else:
        fail_count += 1

print(f"Total Students: {len(marks)}")
print(f"Passing Students (Marks >= {PASS_MARK}): {pass_count}")
print(f"Failing Students (Marks < {PASS_MARK}): {fail_count}")
```

```python
#3

names = ["Alice","Bob","Charlie"]

cleaned_names = [name.strip().lower() for name in names]

print("original list:",names)
print("cleaned list:",cleaned_names)
```

```python
#4

messages =  ["Hii","Welcome to platform","OK"]
LENGTH_LIMIT =10

for msg in messages:
  msg_length =len(msg)

  if msg_length > LENGTH_LIMIT:
    print(f" '{msg}'-> Length:{msg_length} [FLAGGED: Too Long]")
  else:
    print(f" '{msg}'-> Length:{msg_length}")
```

```python
#5

logs= ["INFO","ERROR","WARNING","ERROR"]
error_count = 0
for index,status in enumerate(logs):
  if status == "ERROR":
    print(f"Alert! Error detected at index {index}")
    error_count +=1

    print("---")
    print(f"Scan complete. Total Errors Found :{error_count}")
```
