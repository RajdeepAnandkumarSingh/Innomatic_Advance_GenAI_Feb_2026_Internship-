```python
#1
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

max_score = max(employees.values())

top_performers = [name for name, score in employees.items() if score == max_score]

performers_str = ", ".join(top_performers)
print(f"Top Performers Eligible for Bonus: {performers_str} (Score: {max_score})")
```

```python
#2
import string

query = "Buy mobile phone buy phone online"

query_cleaned = query.lower().translate(str.maketrans('', '', string.punctuation))
keywords = query_cleaned.split()

frequency = {}
for word in keywords:
    frequency[word] = frequency.get(word, 0) + 1

repeated_keywords = {word: count for word, count in frequency.items() if count > 1}

print(repeated_keywords)
```

```python
#3
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = [(index, reading) for index, reading in enumerate(sensor_readings) if reading % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#4
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_counts = {}
for email in emails:
    domain = email.split('@')[1]
    domain_counts[domain] = domain_counts.get(domain, 0) + 1

total_emails = len(emails)

for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
#5
sales = [1200, 1500, 900, 2200, 1400, 3000]

avg_sales = sum(sales) / len(sales)

for i, sale in enumerate(sales, 1):
    if sale >= avg_sales * 1.29:
        print(f"Day {i}: {sale}")
```
